```python
import pandas as pd
```

```python
# Series : 1d labelled(index) array
# Dataframe : 2d labelled matrix like table with labels(index)
# index : numerical val act as positional label
# label : can be any type of val 
# on table(series,dataframe) both can exist and useful to access info
```

```python
# dataframe
# creation using dictionary
df=pd.DataFrame({
    "name":["prashant","amit","sumit","adesh"],
    "cgpa":[6.7,5.6,8.8,9.9]
})

df
```

```python
# customize labels
info={
    "name":["prashant","amit","sumit","adesh"],
    "cgpa":[6.7,5.6,8.8,9.9]
}

df=pd.DataFrame(info,index=["A","B","C","D"])

df
```

```python
# series : homogeneous in nature like numpy 1d array
# Series size is IMMUTABLE , but its values are mutable
# In series vectorized operations are allowed
info=[1,2,6,2,7,3]
s=pd.Series(info)
print(s,type(s))
```

```python
print(s[0:2])
print(s[3])
```

```python
#custom labels
info=[1,2,6,2,7,3]
s=pd.Series(info,index=["A","B","C","D","E","F"])
s
```

```python
print(s["A"]) # accessing through labels
print(s[0]) # accessing through index
print(s.index)
```

```python
# DataFrame creation ways : dictionary , list of lists, numpy 2d arrays
# using dictionary
students={
    "Name":["Prashant","Amit","Sumit","guli","laki","kanchu","shivi"],
    "Age":[20,12,12,15,22,12,8],
    "Class":[10,11,12,7,5,2,6]
}
df=pd.DataFrame(students)
print(df.index,df.columns)
df
```

```python
# using list of lists
students=[["Ram",22],["Amit",23],["sumit",23],["gg",45],["shyam",34]]
df=pd.DataFrame(students,columns=["Name","Age"]) # if we do not specify column names then indices(0,1,...) will be used
print(df.index,df.columns)
df
```

```python
# using numpy arrays
import numpy as np
np_arr=np.array([["amit",98],["sumit",99],["sam",45]])
df=pd.DataFrame(np_arr,columns=["Name","marks"])
print(df.index,df.columns)
df
```
