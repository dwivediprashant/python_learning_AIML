```python
import numpy as np
```

```python
arr=np.array([4,7,8,3,1,7,7,9,6,5])
```

```python
print(np.sum(arr))
print(np.prod(arr))
print(np.std(arr))
print(np.var(arr))
print(np.mean(arr))
print(np.sqrt(arr))
print(np.square(arr))
print(np.pow(arr,0)) # cal power here power=0
print(np.max(arr))
print(np.min(arr))
print(np.argmax(arr)) # return index of max val
print(np.argmin(arr)) # return index of min val
print(np.median(arr))
```

```python
# logarithmic fx
print(np.log(arr)) #base e ie natural log ln
print(np.log10(arr))
print(np.log2(arr))
print(np.exp(arr)) # cal e ki power for each ele
```

```python
 # round fxn
print(np.round(3.16))
print(np.round(3.66))
print(np.ceil(3.66))
print(np.floor(3.66))
print(np.trunc(3.66))
```

```python
# other fxn
print(np.sort(arr))
print(np.unique(arr))
arr=np.array([-3,3,-1,0,7,-7])
print(np.abs(arr))
```
